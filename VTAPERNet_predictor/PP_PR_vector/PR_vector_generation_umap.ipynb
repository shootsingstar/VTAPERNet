```python
import pandas as pd
import umap
```

```python
data = pd.read_csv("PP_740.csv",header = 0,index_col = 'name')
for column in data.columns[1:]:
    data[column] = data[column].astype(float)
data[:5]
```

```python
# Use UMAP to reduce to one dimension
# UMAP involves random initialization and random neighbor search during the dimensionality reduction process.
# These random processes can lead to slightly different results each time UMAP is run.
# By setting a random seed (random_state), you can reduce or eliminate this randomness, ensuring consistent results across runs.
# example:
# np.random.seed(42)
# reducer = umap.UMAP(n_components=1, min_dist=0.1, n_neighbors=5, random_state=42)

reducer = umap.UMAP(n_components=1, min_dist=0.1, n_neighbors=5) 

# Note:
# To ensure that your model uses the same dimensionality reduction results during both training and testing,
# it's important to use the same random seed and UMAP configuration in both phases.
# This avoids variations in the dimensionality reduction that could lead to performance fluctuations or evaluation errors.
```

```python
embedding = reducer.fit_transform(data.values)
embedding.shape
```

```python
#Splice the physical and chemical properties and corresponding embedding values ​​into a DataFrame
df_embedding =  pd.DataFrame(embedding)
df_embedding.columns = ['embedding']

properties = data.index
df_properties =  pd.DataFrame(properties)

df_order = df_properties
df_order.columns = ['properties']
df_order['embedding'] = df_embedding['embedding']

#Sort by embedding column
df_order.sort_values(by='embedding',axis=0,ascending='True', inplace=True)
df_order = df_order.reset_index(drop = True)
```

```python
df_order.to_csv('PR_740.csv')
```
